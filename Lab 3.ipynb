```python
def classify_emails(emails):
    important_keywords = ['urgent', 'important', 'action required', 'deadline']
    promotion_keywords = ['sale', 'discount', 'offer', 'limited time']
    spam_keywords = ['lottery', 'inheritance', 'free', 'click here']

    important_emails = []
    promotion_emails = []
    spam_emails = []

    i = 0
    while i < len(emails):
        lowercase_email = emails[i].lower()
        if any(keyword in lowercase_email for keyword in important_keywords):
            important_emails.append(emails[i])
        elif any(keyword in lowercase_email for keyword in promotion_keywords):
            promotion_emails.append(emails[i])
        elif any(keyword in lowercase_email for keyword in spam_keywords):
            spam_emails.append(emails[i])
        i += 1

    return important_emails, promotion_emails, spam_emails


emails = [
    "URGENT: Action Required by Deadline",
    "Great deals and offers inside! Limited time only!",
    "You've won a free vacation!",
    "Important: Meeting tomorrow",
    "Special discount offer just for you",
    "Claim your inheritance",
    "Click here for exciting opportunities"
]


important_emails, promotion_emails, spam_emails = classify_emails(emails)


print("Important Emails:")
print(important_emails)
print("\nPromotion Emails:")
print(promotion_emails)
print("\nSpam Emails:")
print(spam_emails)
```

```python
import pandas as pd

# Load the dataset
dataset_url = "https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/vgsalesGlobale.csv"
game_data = pd.read_csv(dataset_url)


print("First few rows of the dataset:")
print(game_data.head())


total_global_sales = game_data['Global_Sales'].sum()
print("\nTotal global sales for all games:", total_global_sales)


highest_sales_game = game_data.loc[game_data['Global_Sales'].idxmax()]
print("\nGame with the highest global sales:")
print(highest_sales_game)


average_sales_per_region = game_data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()
print("\nAverage sales for each region:")
print(average_sales_per_region)


top_5_genres_global_sales = game_data.groupby('Genre')['Global_Sales'].sum().nlargest(5)
print("\nTop 5 most popular game genres based on global sales:")
print(top_5_genres_global_sales)


total_sales_per_genre_region = game_data.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
most_popular_genre_na_sales = total_sales_per_genre_region['NA_Sales'].idxmax()
print("\nMost popular genre in North America (NA_Sales):", most_popular_genre_na_sales)


print("\nLast 10 rows of the dataset:")
print(game_data.tail(10))


top_3_platforms_global_sales = game_data.groupby('Platform')['Global_Sales'].sum().nlargest(3)
print("\nTop 3 gaming platforms with the highest global sales:")
print(top_3_platforms_global_sales)


genre_market_share = (game_data.groupby('Genre')['Global_Sales'].sum() / total_global_sales) * 100
print("\nMarket share (percentage of global sales) for each genre:")
print(genre_market_share)


correlation_coefficient = game_data['NA_Sales'].corr(game_data['EU_Sales'])
print("\nCorrelation coefficient between NA_Sales and EU_Sales:", correlation_coefficient)
```
